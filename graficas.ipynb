```python
import numpy as np
import matplotlib.pyplot as plt

# --- Dataset 1: ventas mensuales (12 meses)
np.random.seed(42)  # para reproducibilidad
meses = np.arange(1, 13)          # 1 a 12
ventas = 100 + 20*np.random.randn(12)  # ruido normal alrededor de 100

print("Dataset 1 (ventas mensuales):")
print("Meses:", meses)
print("Ventas:", ventas)
print()

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(meses, ventas, marker="o", linewidth=2, color="blue")
plt.xlabel("Mes")
plt.ylabel("Ventas (unidades)")
plt.title("Ventas por mes - gráfico de línea")

plt.subplot(1, 2, 2)
plt.bar(meses, ventas, color="skyblue", edgecolor="black")
plt.xlabel("Mes")
plt.ylabel("Ventas (unidades)")
plt.title("Ventas por mes - gráfico de barras")

plt.tight_layout()
plt.show()
```

```python
# --- Dataset 2: puntaje de 4 grupos
grupos = ["A", "B", "C", "D"]
puntajes = np.random.randint(60, 100, size=4)

print("Dataset 2 (puntajes por grupo):")
print("Grupos:", grupos)
print("Puntajes:", puntajes)
print()

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.bar(grupos, puntajes, color=["red", "green", "blue", "orange"])
plt.xlabel("Grupo")
plt.ylabel("Puntaje")
plt.title("Puntaje promedio por grupo")

plt.subplot(1, 2, 2)
plt.pie(puntajes, labels=grupos, autopct="%1.1f%%", startangle=90, colors=["red", "green", "blue", "orange"])
plt.title("Distribución de puntajes por grupo")

plt.tight_layout()
plt.show()
```

```python
# --- Dataset 3: nube de puntos (x, y)
np.random.seed(13)
x = np.linspace(0, 10, 100)
y = 2*x + 5 + np.random.randn(100)*3   # y = 2x + 5 + ruido

print("Dataset 3 (nube de puntos):")
print("Tamaño de x:", len(x))
print("Valores de x (primeros 5):", x[:5])
print("Valores de y (primeros 5):", y[:5])
print()

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.scatter(x, y, alpha=0.6, color="purple")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Nube de puntos (x, y)")

plt.subplot(1, 2, 2)
plt.plot(x, y, linestyle="-", marker="o", markersize=3, alpha=0.4, color="purple")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Puntos con línea de tendencia")

plt.tight_layout()
plt.show()
```
